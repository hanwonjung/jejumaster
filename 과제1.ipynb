**우리의 이웃은 누구인가**

목적 : 우리나라에 입국한 외국인들의 수와 국적을 알아보고 현재 우리나라에 체류하고 있는 사람들에 대해서 다시금 생각해 본다.

```python
PATH = './sample_data/'
FILE = '국적_체류자격별_외국인_입국자_월간__연간__202311021620.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file = PATH+FILE
#df = pd.read_csv(file, encoding = 'cp949') #cp949: 윈도우OS 기본 인코딩 포맷
df = pd.read_csv(file, encoding = 'utf-8')
df
```

```python
df.columns
df
```

데이터 프레임


```python
#데이터 프레임 과정
```

```python
#합계 행 삭제
df = df[df.국적별 != '합계']
df
```

```python
#체류자 수로 정렬 (내림차순)
df_my_group = df.sort_values(by=['계'],ascending=False)
df_my_group
```

```python
#2022년 상위30개 국적만 출력
df_my_group_100 = df_my_group.head(30)
df_my_group_100
```

```python
#2022년 하위10개 국적만 출력
df_my_group_100 = df_my_group.tail(10)
df_my_group_100
```

데이터 분석을 통한 정리
최근에 아는 지인의 동생분이 갑자기 조지아라는 나라로 이민을 갔다고 한다. 여행이나 이직도 아닌데 갑자기 새로운 나라에 대한 동경이 생겨서 이민을 신청했다고 해서 놀란 적이 있다. 친동생이 또한 10여년전에 공부를 위해서 독일에서 유학을 한 적이 있다. 한국에 들어오면서 독일에서 지내던 라트비아 친구와 가정이 함께 한국에서 생활할 수 있게 도와주었고 실제로 친정 부모님댁에서 5년정도 함께 살기도 했다. 한국말도 아예 할 수 없었던 그 가정의 두 남매가 한국생활에서 취업이나 학교 생활에서도 매우 잘 적응했는데 바로 그들이 영어를 구사하는 독일 국적을 가진 외국인이였기 때문이다. 외국인 특히 백인 사람이 많이 사는 곳이 아니였기 때문에 사람들의 호기심과 관심 그리고 친절도가 다른 나라 사람들 보다 높았다. 그러나 과연 그들이 선진국이 아닌 다른 나라에서 온 사람들이 였어도 많은 사람들의 관심을 받을 수 있었을까...
 개인적으로 난민, 외국인 노동자, 북한이주민에 대한 관심도 많은 편이고 그들을 부족하지만 꾸준히 다양한 방법으로 후원하고 있는데 코로나 이후 사실 이들에 대한 나의 관심이나 사회의 관심이 더 줄어드는 것을 체감하게 되었다.
 데이터를 통해서 알게 된 것은 아시아, 동남아 지역에서 온 사람들이 많을 것이라고 생각은 했지만 중국에서 온 사람들의 수가 압도적으로 많다는 것이 놀라웠다. 상위 30개 국의 리스트는 어느정도 예상한 나라들이 많았지만 하위 리스트를 보니 정말 한번도 들어보지 못한 나라들이 대부분이였다. 정말 우리나라에 내가 생각하고 들어보지 못한 국적을 가진 사람들이 함께 와서 살고 있다는 사실도 신기했다.
  북한을 이탈해 온 학생들을 5년정도 수업한 적이 있는데 이때 중국 국적을 가진 사람들이 많았었는데 진짜 한족 뿐 아니라 북한에서 자랐지만 중국국적을 가지고 온 사람들이 저 데이터 안에 많이 포함 되어 있지는 않을까 하는 생각이 들었다.
 데이터 분석을 통해서 막연하게나마 추측했던 나의 생각들이 수치화 되어 나타나니 내 옆에 살고 있지만 내가 한번도 이야기 나누지 못했던 나라의 이웃들이 많다는 것도 알게 되었다. 그들을 돕는 존재로 보는 것이 아니라  함께 한국 땅에서 즐겁고 행복한 삶을 누리는 이웃이 되고 싶다.
